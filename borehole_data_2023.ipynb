```python
import gmplot as gp
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import numpy as np
```

```python
df = pd.read_csv('boreholes_2023.csv', header=[0])
# df = pd.read_csv('data_stephen_fix_header.csv', header=[0])
```

```python
# "Materials"

print("Number of Not Nans: ", df.materials.notna().sum())

plt.figure(figsize=(16,6))
b = sb.barplot(np.arange(df.materials.nunique()), df.materials.value_counts())
b.set_xticklabels(df.materials.value_counts().index, rotation=90)

# plt.yticks(np.arange(0,600,50))
plt.grid()
plt.title('Histogram of Materials')
plt.show()
```
